```python
import glob
import numpy as np
import matplotlib.pyplot as plt

filenames = sorted(glob.glob('inflammation*.csv'))
filenames = filenames[0:3]

for filename in filenames:
    print(filename)
    
    data = np.loadtxt(fname=filename, delimiter=',')
    
    fig = plt.figure(figsize=(10.0, 3.0))
    
    axes1 = fig.add_subplot(1, 3, 1)
    axes2 = fig.add_subplot(1, 3, 2)
    axes3 = fig.add_subplot(1, 3, 3)
    
    axes1.set_ylabel('average')
    axes1.plot(np.mean(data, axis=0))
    
    axes2.set_ylabel('max')
    axes2.plot(np.max(data, axis=0))
    
    axes3.set_ylabel('min')
    axes3.plot(np.min(data, axis=0))
    
    fig.tight_layout()
    plt.show()
```

### plot differences

```python
# imports
import glob
import numpy as np
import matplotlib.pyplot as plt

# filenames
filenames = sorted(glob.glob('inflammation*.csv'))

# assign from filenames file 0 and file 1 to variables
data0 = np.loadtxt(fname=filenames[0], delimiter=',')
data1 = np.loadtxt(fname=filenames[1], delimiter=',')

# make a figure
fig = plt.figure(figsize=(10.0, 3.0))

# plot the difference in average
plt.ylabel('Difference in average')
plt.plot(np.mean(data0, axis=0) - np.mean(data1, axis=0))

# tighten up layout
# show, if we need to 
fig.tight_layout()
plt.show()
```

### generate composite statistics

mean across days from all files

```python
# get filenames for all inflammation files with glob
filenames = glob.glob('inflammation*.csv')
 
composite_data = np.zeros((60, 40))

for filename in filenames:
    # sum each file's data into composite_data as it's read
    data = np.loadtxt(filename, delimiter=',')
    composite_data = data + composite_data
    
# and then divide the composite_data by number of samples
# i.e. the number of filenames
# remember: len function
composite_data = composite_data / len(filenames)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(10.0, 3.0))

ax[0].set_ylabel('average')
ax[0].plot(np.mean(composite_data, axis=0))

ax[1].set_ylabel('max')
ax[1].plot(np.max(composite_data, axis=0))

ax[2].set_ylabel('min')
ax[2].plot(np.min(composite_data, axis=0))

fig.tight_layout()
plt.show()
```
